Funciones para el manejo de gráficos con el paquete ggplot2
La función qplot()
La función qplot() se puede utilizar para crear los tipos de gráficos más comunes en ggplot2. Si bien no expone toda la potencia de ggplot, sí permite crear una gama muy amplia de gráficos. El formato es: qplot(x, y, data=, color=, shape=, size=, alpha=, geom=, method=, formula=, facets=, xlim=, ylim= xlab=, ylab=, main=, sub=) Donde, por ejemplo:
alpha: Es la transparencia para los elementos superpuestos expresados como una fracción entre 0 (transparencia completa) y 1 (opacidad completa)
geom: Especifica los objetos geométricos que definen el tipo de gráfico. La opción geom se expresa como un vector de caracteres con una o más entradas. valores geom incluyen “punto”, “diagrama de caja” “suave”, “raya”, “histograma”, “densidad”, “bares”, y “jitter”.

Personalización Gráficos ggplot2.
A diferencia de los gráficos propios de R, los gráficos ggplot2 no se parametrizan mediante la función par(). Los gráficos ggplot2 se pueden modificar mediante la función theme() la cual agrega parámetros a la función qplot(). Tenga en cuenta que las funciones ggplot2 pueden encadenarse con signos “+” para generar el gráfico final, como se ilustra en el siguiente ejemplo:

```r
graf <- qplot(hp, mpg, data=mtcars, shape=am, color=am,facets=gear~cyl, main="Gráfico Millas x Galón vs Caballos de  Fuerza",xlab="Caballos de  Fuerza", ylab="Millas por Galón")

# Agrega fondo blanco y líneas de la cuadrícula de color negro.
graf + theme_bw()
```

```r
# Agrega etiquetas grandes marrones en negrita y cursiva
# La leyenda colocada en la parte superior del gráfico
graf + theme(axis.title=element_text(face="bold.italic",size="12", color="brown"), legend.position="top") 
```

4. La siguiente función en R recibe un DataFrame (DF) y un número de columna (NC) y que retorna en una lista el nombre de la variable correspondiente al número de columna, la media, la mediana, la desviación estándar,la varianza, el máximo y el mínimo:

```r
Estadisticas <- function(DF,NC) {
  media <- mean(DF[,NC])
  mediana <- median(DF[,NC])
  deviacion <- sd(DF[,NC])
  varianza <- var(DF[,NC])
  maximo <- max(DF[,NC])
  minimo <- min(DF[,NC])
  return(list(Variable=colnames(DF)[NC],Media=media,Mediana=mediana,DesEst=deviacion,Varianza=varianza,Maximo=maximo,Minimo=minimo))
}  

setwd("/home/jovyan/work/unidad_tres")
Datos<-read.table("EjemploEstudiantes.csv",header=TRUE,sep=";",dec=",",row.names=1)
Estadisticas(Datos,2)
```

7. Camino Aleatorio (“Random Walk”).
Un camino aleatorio o camino al azar es un proceso estocástico, en el cual la primera diferencia, es un ruido blanco.

```r
w<-rnorm(100)  # 100 números que suguen N(0,1)
x<-w
for(t in 2:100) 
  x[t]<-x[t-1]+w[t] # genera un camino Aleatorio pues 
                    # la primera diferencia es: x[t]-x[t-1]=w[t] (azar)
par(mfrow=c(2,2))
ts.plot(x, main="Camino aleatorio X_t")
acf(x, main="Autocorrelación Simple de Xt",ylim=c(-1,1),col="black")
d<-diff(x)
ts.plot(d,main="Primera diferencia de X_t")
acf(d, main="Autocorrelación Simple de d",ylim=c(-1,1),ci.col="black")
```
