#### Una persona desea sacar un préstamo, de P colones a una tasa de interés mensual i. El préstamo tiene que ser reembolsado en n cuotas mensuales de tamaño R, comenzando dentro de un mes. El problema es calcular R. La fórmula R es:

#### $$R = P  \Bigg (\frac{i}{1-\frac{1}{(1+i)^{n}} }\Bigg)$$

#### Si supongamos que P=150,000 que la tasa de interés es del 2% y que le número de pagos es 10. El código en R sería:


```r
tasa.interes <- 0.02
n <- 10000
principal <- 150000
pago <- principal * tasa.interes / (1 - (1 + tasa.interes)^(-n))
pago
```

Sería mejor escribir una función en R para este cálculo como sigue:


```r
calcula.cuota <- function(tasa.interes,n,principal) {
   pago <- principal * tasa.interes / (1 - (1 + tasa.interes)^(-n))
   return(pago)
}

calcula.cuota(0.02,10,150000)
```

```r
calcula.cuota(0.11,340,3500000)
```

#### Función para calcular el factorial de n que se define como: n = 1 * 2 * 3 * n

```r
factorial <- function(n) {
    fact <- prod(1:n)
    return(fact)
}

factorial(9)
```

#### Función para calcular las combinaciones r en n que se define como:

$$\binom{n}{r} = \frac{n!}{r!(n-r)!}$$


```r
combinaciones <- function(r, n) {
    comb <- factorial(n)/(factorial(r)*factorial(n-r))
    return(comb)
}

combinaciones(3,9)
```

#### En una función el orden de los parámetros puede cambiarse


```r
combinaciones(3,9) # Orden por defecto
```

```r
combinaciones(n=9,r=3) # Para cambiar el orden se especifica el nombre del parámetro
```

```r
combinaciones(9,3)  # Da error por que r debe ser menor a n
```

#### Función R que recibe un par (x,y) y retorna el cuadrante en donde se encuentra el punto.


```r
cuadrante <- function(x,y) {
  if (x > 0) {
    if(y > 0) {
      r<-1
    }  
    else {
      r<-4
    }  
  }
  else {
    if(y > 0) {
      r<-2
    }  
    else {
      r<-3
    }  
  }
  return(r)
}

cuadrante(4,3)
```

```r
cuadrante(-6,1)
```

#### Este algoritmo dados tres valores almacenarlos en las variables A, B y C retorna cual es el mayor.


```r
mayor <- function(A,B,C) {
  if((A > B) && (A > C)) {
    return(A)
  }
  else {
    if((B > A) && (B > C)) {
      return(B)
    }
    else {
      return(C)  
    }  
  }  
}

mayor(9,0,2)
```

```r
mayor(0,-6,-3)
```

```r
mayor(1,2,2)
```

#### Función R que construye un vector “v” tal que en la entrada k está el valor k ** 2 - 4, es decir,

$$v[k] = k^2 - 4$$


```r
cvector.f <- function(n) {
  v<-rep(0,n) # Inicializa un vector con n ceros
  for(k in 1:n) {
    v[k]<-k^2-4
  }
  return(v)
}

cvector.f(9)
```

```r
cvector.f(factorial(10))
```

#### La siguiente función calcula la suma del valor absoluto de las entradas de un vector, es de decir, la norma 1 del vector dada por |v[1]| + |v[2]| + . . . + | v[n] | :


```r
norma1.f <- function(v) {
   suma <- 0
   for(i in 1:length(v)) {
      suma <- suma + abs(v[i])
   }
   return(suma)
}

vec0<-c(4,-7,2,1)
norma1.f(vec0)
```

#### La siguiente función encuentra la primera pocisión en un vector en la que se encuentra en valor 0. La siguiente es una versión con “for”.


```r
encuentra.cero.f <- function(v) {
  s <- -1
  for(i in 1:length(v)) {
    if((v[i]==0) && (s == -1)) {
      s<-i
    }
  }
  if (s != -1) {
    return(s)
  }
  else {
    return("¡El vector no tiene ningún cero!")
  }
}

vec<-c(4,-7,2,1,9)
encuentra.cero.f(vec)
```

```r
vec<-c(4,-7,2,0,9,98,0,9,-1)
encuentra.cero.f(vec)
```

#### La siguiente función multiplica una matriz por un vector, es decir, calcula y = Ax donde A es una matriz de tamaño m × n, “x” es un vector (matriz) tamaño n × 1 y por lo tanto “y” es un vector (matriz) tamaño m × 1.


```r
mat.x.vec <- function(A,x) {
  m <- nrow(A)
  n <- ncol(A)
  y <- matrix(0,nrow=m)
  for (j in 1:n) {
    for (i in 1:m) {
      y[i] <- y[i] + A[i,j]*x[j]
    }
  }
  return(y)
}

M <- matrix(c(2,3,-4,4,-8,2,0,1,-23,-2,4,0),4,3)
M
```

```r
v <- matrix(c(1,3,3),3,1)
v
```

```r
res<-mat.x.vec(M,v)
res
```

#### La siguiente función verifica que dos matrices son del mismo tamaño.


```r
mismo.tamano <- function(A,B) {
  if((nrow(A)==nrow(B)) && (ncol(A)==ncol(B))) {
    return(TRUE)
  }
  else {
    return(FALSE)
  }
}

M1 <- matrix(c(2,3,-4,4,-8,2,0,1,-23,-2,4,0),4,3)
M1
```

```r
M2 <- matrix(c(0,0,0,1,1,2,0,1,-6,-2,1,0),4,3)
M2
```

```r
mismo.tamano(M1,M2)
```

```r
M3 <- matrix(c(1,1,2,0,1,-6,-2,1,0),3,3)
M3
```

```r
mismo.tamano(M1,M3)
```

#### La siguiente función suma dos matrices del mismo tamaño m×n.


```r
# Esta función asume que A y B son del mismo tamaño
# Las validaciones nunca se hacen dentro de una función
suma.mat <- function(A,B) {
  m<-nrow(A)
  n<-ncol(A)
  C <- matrix(0,nrow=m,ncol=n)
  for (j in 1:n) {
    for (i in 1:m) {
      C[i,j] <- A[i,j]+B[i,j]
    }
  }
  return(C)
}

M1 <- matrix(c(2,3,-4,4,-8,2,0,1,-23,-2,4,0),4,3)
M1
```

```r
 M2 <- matrix(c(0,0,0,1,1,2,0,1,-6,-2,1,0),4,3)
M2
```

```r
# Esto es una validación, las validaciones nunca se hacen dentro de una función
if(mismo.tamano(M1,M2)==TRUE) {
  M3<-suma.mat(M1,M2)
  M3
}  
```

#### La siguiente función en R recibe un DataFrame (DF) y retorna el valor máximo almacenado en DF

```r
maximo <- function(DF) {
  n<-dim(DF)[1]
  m<-dim(DF)[2]
  max<-DF[1,1]
  for(i in 1:n) {
    for(j in 1:m) {
      if(DF[i,j]>max) {
        max<-DF[i,j]
      }
    }
  }
  return(max)
}  

maximo(Datos)
```

#### La siguiente función en R recibe un DataFrame (DF) y retorna en una lista el valor mímimo, el máximo, la cantidad de ceros y la cantidad de números pares que contiene el dataFrame:


```r
valores <- function(DF) {
  n<-dim(DF)[1]
  m<-dim(DF)[2]
  min<-DF[1,1]
  max<-DF[1,1]
  t.ceros<-0
  t.pares<-0
  for(i in 1:n) {
    for(j in 1:m) {
      if(DF[i,j]>max) {
        max<-DF[i,j]
      }
      if(DF[i,j]<min) {
        min<-DF[i,j]
      }
      if(DF[i,j]==0) {
        t.ceros<-t.ceros+1
      }
      if((DF[i,j]%%2)==0) {
        t.pares<-t.pares+1
      }
    }
  }
  return(list(Maximo=max,Minimo=min,Numero.Ceros=t.ceros,Pares=t.pares))
}  
## Vamos a establecer el directorio
setwd("/home/jovyan/work/tratamiento")
Datos<-read.table("EjemploEstudiantes.csv",header=TRUE,sep=";",dec=",",row.names=1)
valores(Datos)
```

#### La siguiente función en R recibe un DataFrame (DF) y un número de columna (NC) y que retorna en una lista el nombre de la variable correspondiente al número de columna, la media, la mediana, la desviación estándar,la varianza, el máximo y el mínimo:


```r
Estadisticas <- function(DF,NC) {
  media <- mean(DF[,NC])
  mediana <- median(DF[,NC])
  deviacion <- sd(DF[,NC])
  varianza <- var(DF[,NC])
  maximo <- max(DF[,NC])
  minimo <- min(DF[,NC])
  return(list(Variable=colnames(DF)[NC],Media=media,Mediana=mediana,DesEst=deviacion,Varianza=varianza,Maximo=maximo,Minimo=minimo))
}  
Estadisticas(Datos,2)
```
