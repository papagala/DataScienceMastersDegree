6.	Utilice la tabla de datos de Compras.csv, realice 2 gráficos con ggplot2,  explique la interpretación del gráfico y los parámetros de la función ggplot.

```r
setwd("/home/jovyan/work/unidad_tres")
compras <- read.csv("Compras.csv",row.names=1)
```

```r
compras
```

### Dado que no tengo suficiente información sobre que significa la columna Juguetes (número de juguetes?) y Hogar (número identificador de una casa hogar?), Agruparé por Número de Juguetes para visualizar de maner más sencilla

```r
sum_juguetes_genero <- aggregate(Juguetes ~ Tienda + Genero, compras, FUN=sum)
sum_juguetes_genero
```

### Una vez agrupado por número de juguetes y género, usaré una gráfica de barras (stacked en Excel). Los parametros son:

- data = el DataFrame con los datos a graficar
- aes = mapeo estético que describe como las variables en los datos están mapeados a propiedades visuales de las geometrías. 
- geom_bar hace que la altura de la barra sea proporcional al número de casos de cada grupo

### Aquí los grupos horizontales son las tiendas y la altura de las barras es el número de juegues. Existe además una dimensión extra para el género en donde entra el color de la barra


```r
# Gráfico básico
ggplot(data=sum_juguetes_genero, aes(x=Tienda, y=Juguetes, fill=Genero)) + geom_bar(stat="identity")
```

### Aquí agrupamos hogar por género y tienda

```r
sum_juguetes_hogar <- aggregate(Hogar ~ Genero + Tienda, compras, FUN=sum)
sum_juguetes_hogar
```

### Aquí son los mismos parámetros de arriba pero ahora tenemos el género en el eje x, la altura es el número de hogar por género y de colores tenemos las tiendas 

```r
# Gráfico básico
ggplot(data=sum_juguetes_hogar, aes(x=Genero, y=Hogar, fill=Tienda)) + geom_bar(stat="identity")
```
